```python
import pandas as pd

# 读取excel的内容并打印
df = pd.read_excel(".//res//ai绘画 - 小红书搜索-采集的数据-后羿采集器.xlsx")
# print(df)
# 打印所有列的数据
print(df.columns)
# 打印count列
# print(df['count'])
# 删除重复行的数据并保存
# print(df['count'].unique())
# 保存到df中


# 去重后的数据
df = df.drop_duplicates()
print(df)
# print(df['count'].drop_duplicates())
```

```python
# 对该列数据进行归一化处理，需要进行判断，若字符最后有w,如1.1w应去掉w和.,再加000
# print(df['count'])
df['count'] = df['count'].map(lambda x: x.replace('w', '').replace('.', '') + '000' if 'w' in x else x)
print(df['count'])
# 把count列的数据类型转换为int
df['count'] = df['count'].astype(int)
# 对count列进行排序
df = df.sort_values(by='count', ascending=False)
# 提取前20条数据的title和count列
df = df[['标题', 'count']][:20]
# 将数据写入excel
df.to_excel('.//res//ai绘画 - 小红书搜索-采集的数据-后羿采集器-处理后.xlsx', index=False)
print(df)
```
