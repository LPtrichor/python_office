```python
import pandas as pd

# 从./gpt_accounts.txt中读取文本信息，然后将文本信息分割成行，然后将行号作为键，文本信息作为值，存储到字典中
with open('./0204.txt', 'r') as f:
    text = f.read()
# print(df)
# 打印所有列的数据
# print(df.columns)
# 打印count列
# print(df['count'])
# 删除重复行的数据并保存

# 分割文本信息
lines = text.strip().split('\n')
# 每一行通过----分割成5个部分 0:账号 1:密码 2:密码 3:密码 4:session 5:key
lines = [line.split('----') for line in lines]
# 打印lines
print(lines)
```

```python
for line in lines:
    print(line[2])
```

## 处理和归一化

```python
# 有些行是5列，有些行是4列
# 对于5列行其中第二三列是重复的，直接删除
# 统一为四列
lines = [line[:2] + line[3:] if len(line) == 5 else line for line in lines]
# 打印lines
print(lines)
```

```python
# 将lines转换为字典 存储到gpt_accounts的json文件中
df = pd.DataFrame(lines)
# 打印df
print(df)
df.columns = ['account', 'password', 'session', 'key']
# 将df转换为字典
df = df.to_dict(orient='index')
# df的行号不需要，所以删除
# df = {int(k): v for k, v in df.items()}
# 打印df
print(df)
# 将df保存到gpt_accounts.json文件中
with open('./gpt_accounts.json', 'w') as f:
    f.write(str(df))
```
